```python
#Convolution Function
import numpy as np

def conv(img, fil):
    output = np.zeros(img.shape)
    for i in range(3):
        image = img[:,:,i]
        inv_fil = np.flipud(np.fliplr(fil))
        #print(image.shape)
        final_img = np.zeros(image.shape)
        pad_img = np.zeros((image.shape[0] + 2, image.shape[1] + 2))
        #print(pad_img.shape)
        pad_img[1:-1, 1:-1] = image


        for x in range(image.shape[1]):
            for y in range(image.shape[0]):

                final_img[y, x] = (inv_fil * pad_img[y: y+3, x: x+3]).sum()
        output[:,:,i] = final_img

    return output

from PIL import Image
hx = [[-1,-2,-1],[0,0,0],[1,2,1]]
hy = [[-1,0,1],[-2,0,2],[-1,0,1]]

image = Image.open('cat.jpg')

Gx = conv(np.array(image),np.array(hx))
Gy = conv(np.array(image),np.array(hy))
Gm = np.abs(Gx)+np.abs(Gy)
T = np.average(Gm)
# Gm[Gm>T] = 1
# Gm[Gm<T] = 0
edge_img=np.where(Gm <= T, 0, 1)

## PLOTTER ##
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 7))


fig.add_subplot(3, 2, 1)
plt.imshow(image)
plt.title('Original')
plt.axis('off')
fig.add_subplot(3, 2, 2)
plt.imshow(Gx)
plt.title('Gx')
plt.axis('off')
fig.add_subplot(3, 2, 3)
plt.imshow(Gy)
plt.title('Gy')
plt.axis('off')
fig.add_subplot(3, 2, 4)
plt.imshow(Gm/255)
plt.title('Gm')
plt.axis('off')
fig.add_subplot(3, 2, 5)
plt.title('Edge Image')
plt.imshow(edge_img/255)
plt.axis('off')
plt.imsave('result.jpg',edge_img.astype(np.uint8)/255)
```
